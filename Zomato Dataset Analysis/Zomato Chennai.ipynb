```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "svg"
```

# Q1) How many restaurants are in Chennai?

```python
top_locations = zomato_df['Location'].value_counts().head(20).index
filtered_df = zomato_df[zomato_df['Location'].isin(top_locations)]

fig = px.histogram(
    filtered_df,
    x='Location',
    color='Location',
    title='Top 20 Locations with Most Restaurants in Chennai',
    labels={'Location': 'Location'}
)
fig.show()
```

# Q2) Which are the Top 10 highest rated Dining Biriyani Restaurant in Chennai?

```python
Biriyani_df = zomato_df[zomato_df['Cuisine'].str.contains('Biryani')]
Biriyani_df.sort_values(by='Dining Rating',ascending=False).head(10)
```

# Q3) Which places have the highest rated Delivery restaurant in Chennai?

```python
highest_rated_df = zomato_df[zomato_df['Delivery Rating'] >= 4.5]
highest_rated_df
```

```python
fig = px.histogram(highest_rated_df, x='Name of Restaurant', color='Location',
title= 'No. of Best Restaurant by Places').update_xaxes(categoryorder="total descending")
fig.show()
```

# Q4) What is the Avg Price for 2 of highest rated restaurant in Chennai?

```python
highest_rated_price_df = highest_rated_df.groupby(by=['Location', 'Name of Restaurant'])['Price for 2'].mean().reset_index()
highest_rated_price_df.head()
```

```python
fig = px.scatter(highest_rated_price_df, x="Location", y="Price for 2", color="Name of Restaurant", symbol="Name of Restaurant",
title=' Avg Price for 2 of High rated restaurant').update_traces(marker_size=10)
fig.show()
```

# Q5) Which areas have a large number of Chinese Restaurant Market?

```python
chinese_df = zomato_df[zomato_df['Cuisine'].str.contains('Chinese')]
chinese_df
```

```python
chinese_rest_df = chinese_df.groupby(by='Location').agg({
    'Name of Restaurant': 'count',
    'Price for 2': 'mean'
}).rename(columns={'Name of Restaurant': 'COUNT OF RESTAURANTS'})

chinese_rest_df = chinese_rest_df.sort_values('COUNT OF RESTAURANTS', ascending=False).head(25)
chinese_rest_df.head()
```

```python
chinese_rest_df = chinese_rest_df.reset_index()
```

```python
fig = px.bar(chinese_rest_df, x='Location', y='COUNT OF RESTAURANTS', color='Price for 2', title= 'No. of Chinese Restaurant by Places')
fig.show()
```
